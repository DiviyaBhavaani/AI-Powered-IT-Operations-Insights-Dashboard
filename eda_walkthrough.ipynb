# EDA Walkthrough

This notebook explores the synthetic incident dataset used by the dashboard project.

```
import pandas as pd

df = pd.read_csv('../data/incidents.csv', parse_dates=['created_at', 'resolved_at'])
df.head()
```

```
df.shape, df['status'].value_counts()
```

```
closed = df[df['status'] == 'Closed']
closed.groupby('category')['breached_sla'].mean().sort_values(ascending=False)
```

```
closed.groupby('severity')['resolution_hours'].mean().reindex(['P1','P2','P3','P4'])
```

```
df[df['status']=='Open'].groupby('assigned_team').size().sort_values(ascending=False).head(10)
```
